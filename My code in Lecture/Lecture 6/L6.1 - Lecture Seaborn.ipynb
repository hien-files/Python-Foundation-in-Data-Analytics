````python
import pandas as pd
import seaborn as sns
````

**A beginning code block for visualization:**

```
sns.barplot(data = dataframe,
            x = 'column_1',
            y = 'column_2')
```
Plotting from dataframe, with column_1 as x, column_2 as y.

By default, if no x and y is defined, the columns name --> x_axis, the index --> y_axis.

````python
df = sns.load_dataset("titanic")
df.head()
````

### sns.countplot()
**Show the counts of observations in each categorical bin using bars.**

Detail: https://seaborn.pydata.org/generated/seaborn.countplot.html


````python
# 👉 Viz Số lượng hành khách trên mỗi class như thế nào?
# Option 1:
sns.countplot(data=df,
              x='class')
````

#### **Make some configs**
##### 👉 Change chart's color:

````python
# Make some configs:
sns.countplot(
            data = df,
            x = df['class'], 
            color= 'orange'
)

# We can add color = "name_color" to change the graph color.
# Also, orient to change the orientation of the graph.
# Change orders
````

### sns.barplot 

https://seaborn.pydata.org/generated/seaborn.barplot.html

````python
df.head()
````

````python
# Example 1: Viz giá vé trung bình trên các class?
# Step 1: Tạo dataframe để vẽ biểu đồ:
plot_data = df.groupby('class')['fare'].mean().reset_index()
plot_data
````

````python
# Step 2: Tiến hành vẽ biểu đồ:
sns.barplot(data=plot_data,
            x = 'class',
            y = 'fare')
````

````python
# Example 2: Viz Tỉ lệ sống xót của giới tính và tình trạng độc thân của họ như thế nào?
dfplot = df.groupby(['sex', 'alone'])['survived'].mean().reset_index()
dfplot
````

````python
# Viz dfplot:
sns.barplot(
    data = dfplot,
    x = 'sex',
    y = 'survived',
    hue = 'alone'
)
````

### sns.lineplot
Line graphs are most commonly used to plot continuous data.Because the points are physically connected via the line, it implies a connection between the points that may not make sense for categorical data (a set of data that is sorted or divided into different categories). Often, our continuous data is in some unit of time: days, months, quarters, or years.

https://seaborn.pydata.org/generated/seaborn.lineplot.html

### sns.boxplot

https://seaborn.pydata.org/generated/seaborn.boxplot.html

<img src="https://help.ezbiocloud.net/wp-content/uploads/2020/04/%E1%84%89%E1%85%B3%E1%84%8F%E1%85%B3%E1%84%85%E1%85%B5%E1%86%AB%E1%84%89%E1%85%A3%E1%86%BA-2020-04-19-%E1%84%8B%E1%85%A9%E1%84%8C%E1%85%A5%E1%86%AB-9.41.35.png" width="600">

Boxplot thể hiện các phân phối dữ liệu, nghĩa là giúp chúng ta biết được độ dàn trải của các điểm dữ liệu như thế nào, dữ liệu có đối xứng không, phân bố rộng hay hẹp, giá trị nhỏ nhất, lớn nhất và các điểm ngoại lệ (outliers)

<img src="https://api.toploigiai.vn/storage/uploads/cach-nhan-xet-bieu-do-hop_3" width="300">

````python
# Calculate the quantiles -- Tính Q1 và Q3
q1 = df['age'].quantile(0.25) # Tính Q1
q3 = df['age'].quantile(0.75) # Tính Q3 

# Calculate the interquantile range
iqr = q3 - q1

# Calculate the whisker
upper = q3 + 1.5*iqr
lower = q1 - 1.5*iqr

print('q1', q1)
print('q3', q3)
print('upper', upper)
print('lower', lower)
````

````python
# Visualize the distribution of age
sns.boxplot(data=df,
            x='age')
````
