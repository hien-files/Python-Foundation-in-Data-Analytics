### sns.boxplot

Boxplot thá»ƒ hiá»‡n cÃ¡c phÃ¢n phá»‘i dá»¯ liá»‡u, nghÄ©a lÃ  giÃºp chÃºng ta biáº¿t Ä‘Æ°á»£c Ä‘á»™ dÃ n tráº£i cá»§a cÃ¡c Ä‘iá»ƒm dá»¯ liá»‡u nhÆ° tháº¿ nÃ o, dá»¯ liá»‡u cÃ³ Ä‘á»‘i xá»©ng khÃ´ng, phÃ¢n bá»‘ rá»™ng hay háº¹p, giÃ¡ trá»‹ nhá» nháº¥t, lá»›n nháº¥t vÃ  cÃ¡c Ä‘iá»ƒm ngoáº¡i lá»‡ (outliers)

<img src="https://api.toploigiai.vn/storage/uploads/cach-nhan-xet-bieu-do-hop_3" width="300">

````python
# Calculate the quantiles -- TÃ­nh Q1 vÃ  Q3
q1 = df['age'].quantile(0.25) # TÃ­nh Q1
q3 = df['age'].quantile(0.75) # TÃ­nh Q3 

# Calculate the interquantile range
iqr = q3 - q1

# Calculate the whisker
upper = q3 + 1.5*iqr
lower = q1 - 1.5*iqr

print('q1', q1)
print('q3', q3)
print('upper', upper)
print('lower', lower)
````

````python
# Visualize the distribution of age
sns.boxplot(data=df,
            x='age')
````
